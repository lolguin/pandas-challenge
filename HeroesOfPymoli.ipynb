```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
csvfile = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(csvfile)

purchase_df.head()
```

```python
# ****CREATE PURCHASING ANLYSIS****

#calculate total players
total_players = len(purchase_df["SN"].unique())

#calculate unique items
unique_items = len(purchase_df["Item ID"].unique())

#calculate average price
average_price = purchase_df["Price"].mean()

#calculate total purchases
total_purchases = len(purchase_df["Purchase ID"].unique())

#calculate total revenue
total_revenue = purchase_df["Price"].sum()

#load these values into a purchase analysis data frame
purchase_analysis_df = pd.DataFrame({
    "Number of Unique Items": [unique_items], "Average Price": [average_price], "Number of Purchases": [total_purchases], "Total Revenue": [total_revenue]
})

#Change format of Average Price and Total Revenue using mapping
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:.2f}".format)

#print purchase analysis df
purchase_analysis_df
```

```python
#create new df with screen name index
gender2_df = purchase_df.groupby("Gender")["SN"].nunique()

gender2_df
```

```python
# ****CREATE GENDER DEMOGRAPHICS****

#create new df with unique screen name
unique_sn = purchase_df.groupby("Gender")["SN"].nunique()

#calculate total number of records
total_records = int(len(purchase_df["SN"].unique()))

#calculate total males, females and others
unique_male_count = int(unique_sn.loc["Male"])
unique_female_count = int(unique_sn.loc["Female"])
unique_other_count = int(unique_sn.loc["Other / Non-Disclosed"])

#create gender dataframe
gender_df = pd.DataFrame({
    "Gender": ["Male", "Female", "Other / Non-Disclosed"], "Total": [unique_male_count, unique_female_count, unique_other_count], "Percentage": [unique_male_count/total_records, unique_female_count/total_records, unique_other_count/total_records]
})

#change Percentage column formatting
gender_df["Percentage"] = gender_df["Percentage"].map("{:.2%}".format)

#print gender data frame
gender_df



```

```python
# ****CREATE PURCHASING ANALYSIS BY GENDER****

male_analysis = purchase_df.loc[purchase_df["Gender"] == "Male", :]

male_average = male_analysis["Price"].mean()
male_count = male_analysis["Price"].count()
male_sum = male_analysis["Price"].sum()
male_avg_total = male_sum/unique_male_count

female_analysis = purchase_df.loc[purchase_df["Gender"] == "Female", :]

female_average = female_analysis["Price"].mean()
female_count = female_analysis["Price"].count()
female_sum = female_analysis["Price"].sum()
female_avg_total = female_sum/unique_female_count

other_analysis = purchase_df.loc[purchase_df["Gender"] == "Other / Non-Disclosed", :]

other_average = other_analysis["Price"].mean()
other_count = other_analysis["Price"].count()
other_sum = other_analysis["Price"].sum()
other_avg_total = other_sum/unique_other_count

#Create 4 seperate lists for summary dataframe
average = [male_average, female_average, other_average]
count = [male_count, female_count, other_count]
sum = [male_sum, female_sum, other_sum]
avg_total = [male_avg_total, female_avg_total, other_avg_total]
```

```python
purchase_analysis = pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"], 
                                  "Purchase Count" : count, 
                                  "Avg Purchase Price" : average, 
                                  "Total Purchase Value": sum, 
                                  "Avg Total Purchase Per Person": avg_total
                                 })

purchase_analysis["Avg Purchase Price"] = purchase_analysis["Avg Purchase Price"].map("${:.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis["Avg Total Purchase Per Person"] = purchase_analysis["Avg Total Purchase Per Person"].map("${:.2f}".format)

purchase_analysis
```

```python
#Age Demographics

bins = [-1, 9,14, 19, 24, 29, 34, 39, 100]
bin_categories = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_summary = purchase_df.drop_duplicates(subset = "SN")

age_summary["Age Bins"] = pd.cut(age_summary["Age"], bins, labels = bin_categories)

age_summary = age_summary.groupby("Age Bins")

age_count = age_summary["Age"].count()

age_summary_table = pd.DataFrame({"Count of Age": age_count, "Percentage": age_count/total_players})

age_summary_table["Percentage"] = age_summary_table["Percentage"].map("{:.2%}".format)

age_summary_table



```

```python
#Purchasing Analysis (Age)
purchase_summary = purchase_df

purchase_summary["Age Bins"] = pd.cut(purchase_summary["Age"], bins, labels = bin_categories)

purchase_summary = purchase_summary.groupby("Age Bins")

purchase_count = purchase_summary["Purchase ID"].count()
average_purchase_price = purchase_summary["Price"].mean()
total_purchase_price = purchase_summary["Price"].sum()
avg_total_purchase = total_purchase_price / age_count

purchase_table = pd.DataFrame({
                "Purchase Count" : purchase_count, 
                "Average Purchase Price" : average_purchase_price, 
                "Total Purchase Value": total_purchase_price,
                "Avg Total Purchase per Person": avg_total_purchase})

purchase_table["Average Purchase Price"] = purchase_table["Average Purchase Price"].map("${:.2f}".format)
purchase_table["Total Purchase Value"] = purchase_table["Total Purchase Value"].map("${:.2f}".format)
purchase_table["Avg Total Purchase per Person"] = purchase_table["Avg Total Purchase per Person"].map("{:,.2f}".format)

purchase_table
```
